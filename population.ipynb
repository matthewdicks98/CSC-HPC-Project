```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## comparing population sizes

```python
mpi_population = pd.read_csv("Results/mpi/mpi_population.txt")
mpi_population[' elapsed/gen_count'] = mpi_population[' elapsed/gen_count']*1000

omp_population = pd.read_csv("Results/omp_crossover/omp_crossover_population")
omp_population[' elapsed/gen_count'] = omp_population[' elapsed/gen_count']*1000

ser_population = pd.read_csv("Results/serial/ser_population")
ser_population[' elapsed/gen_count'] = ser_population[' elapsed/gen_count']*1000
```

```python
mpi_population
```

```python
omp_population
```

```python
ser_population
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(ser_population[' pop_size'],ser_population[' elapsed/gen_count'],marker='o', color="teal")
plt.plot(omp_population[' pop_size'][:7],omp_population[' elapsed/gen_count'][:7],marker='o', color="violet")
plt.plot(mpi_population[' pop_size'][:7], mpi_population[' elapsed/gen_count'][:7],marker='o', color="tab:red")
plt.legend(['serial','openmp','mpi'])
plt.ylabel('Time per 1000 generations')
plt.xlabel('Population Size')
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(omp_population[' pop_size'][:10],omp_population[' elapsed/gen_count'][:10],marker='o', color="violet")
plt.plot(mpi_population[' pop_size'][:10], mpi_population[' elapsed/gen_count'][:10],marker='o', color="tab:red")
plt.legend(['openmp','mpi'])
plt.ylabel('Time per 1000 generations')
plt.xlabel('Population Size')
```
