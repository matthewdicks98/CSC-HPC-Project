```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## comparing population sizes

```python
mpi_population = pd.read_csv("Results/mpi/mpi_population.txt")
mpi_population[' elapsed/gen_count'] = mpi_population[' elapsed/gen_count']*1000

omp_population = pd.read_csv("Results/omp_crossover/omp_crossover_population")
omp_population[' elapsed/gen_count'] = omp_population[' elapsed/gen_count']*1000

ser_population = pd.read_csv("Results/serial/ser_population")
ser_population[' elapsed/gen_count'] = ser_population[' elapsed/gen_count']*1000
```

```python
laptop_mpi_population = pd.read_csv("Results/laptop/mpi/mpi_population.txt")
laptop_mpi_population[' elapsed/gen_count'] = laptop_mpi_population[' elapsed/gen_count']*1000

laptop_omp_population = pd.read_csv("Results/laptop/omp_crossover/omp_crossover_population.txt")
laptop_omp_population[' elapsed/gen_count'] = laptop_omp_population[' elapsed/gen_count']*1000

laptop_ser_population = pd.read_csv("Results/laptop/serial/ser_population.txt")
laptop_ser_population[' elapsed/gen_count'] = laptop_ser_population[' elapsed/gen_count']*1000
```

```python
laptop_mpi_population
```

```python
laptop_omp_population
```

```python
laptop_ser_population
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(ser_population[' pop_size'],ser_population[' elapsed/gen_count'],marker='D', color="teal")
plt.plot(omp_population[' pop_size'][:12],omp_population[' elapsed/gen_count'][:12],marker='o', color="violet")
plt.plot(mpi_population[' pop_size'][:12], mpi_population[' elapsed/gen_count'][:12],marker='x', color="tab:red")
plt.legend(['serial','openmp','mpi'])
plt.ylabel('Time per 1000 generations (μs)')
plt.xlabel('Population Size')
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(omp_population[' pop_size'][:10],omp_population[' elapsed/gen_count'][:10],marker='o', color="violet")
plt.plot(mpi_population[' pop_size'][:10], mpi_population[' elapsed/gen_count'][:10],marker='x', color="tab:red")
plt.legend(['openmp','mpi'])
plt.ylabel('Time per 1000 generations (μs)')
plt.xlabel('Population Size')
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(ser_population[' pop_size'][:6],ser_population[' elapsed/gen_count'][:6],marker='D', color="teal")
plt.plot(omp_population[' pop_size'][:6],omp_population[' elapsed/gen_count'][:6],marker='o', color="violet")
plt.plot(mpi_population[' pop_size'][:6], mpi_population[' elapsed/gen_count'][:6],marker='x', color="tab:red")

plt.plot(laptop_ser_population[' pop_size'],laptop_ser_population[' elapsed/gen_count'],marker='D', color="teal", dashes=[2,2])
plt.plot(laptop_omp_population[' pop_size'][:6],laptop_omp_population[' elapsed/gen_count'][:6],marker='o', color="violet", dashes=[2,2])
plt.plot(laptop_mpi_population[' pop_size'][:6], laptop_mpi_population[' elapsed/gen_count'][:6],marker='x', color="tab:red", dashes=[2,2])

plt.legend(['serial (hpc)','openmp (hpc)','mpi (hpc)','serial (laptop)','openmp (laptop)','mpi (laptop)'])
plt.ylabel('Time per 1000 generations (μs)')
plt.xlabel('Population Size')
```
