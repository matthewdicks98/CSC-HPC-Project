```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## OMP Particle Experiments

```python
omp_fitness = pd.read_csv("Results/omp_functions/C/omp_fitness_particles")
omp_fitnessB = pd.read_csv("Results/omp_functions/B/omp_fitness_particles")
omp_fitness[' elapsed/gen_count'] += omp_fitnessB[' elapsed/gen_count']
omp_fitness[' elapsed/gen_count'] = omp_fitness[' elapsed/gen_count']/2
omp_fitness[' elapsed/gen_count'] = omp_fitness[' elapsed/gen_count']*1000


omp_breeding = pd.read_csv("Results/omp_functions/C/omp_breeding_particles")
omp_breedingB = pd.read_csv("Results/omp_functions/B/omp_breeding_particles")
omp_breeding[' elapsed/gen_count'] += omp_breedingB[' elapsed/gen_count'] 
omp_breeding[' elapsed/gen_count'] = omp_breeding[' elapsed/gen_count']/2
omp_breeding[' elapsed/gen_count'] = omp_breeding[' elapsed/gen_count']*1000


omp_crossover = pd.read_csv("Results/omp_functions/C/omp_crossover_particles")
omp_crossoverB = pd.read_csv("Results/omp_functions/B/omp_crossover_particles")
omp_crossover[' elapsed/gen_count'] += omp_crossoverB[' elapsed/gen_count']
omp_crossover[' elapsed/gen_count'] = omp_crossover[' elapsed/gen_count']/2
omp_crossover[' elapsed/gen_count'] = omp_crossover[' elapsed/gen_count']*1000
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(omp_fitness[' num_particles'][:5], omp_fitness[' elapsed/gen_count'][:5],marker='o', color="forestgreen")
plt.plot(omp_breeding[' num_particles'][:5],omp_breeding[' elapsed/gen_count'][:5],marker='o', color="cornflowerblue")
plt.plot(omp_crossover[' num_particles'][:5],omp_crossover[' elapsed/gen_count'][:5],marker='o', color="violet")
plt.legend(['fitness','fitness and breeding','fitness and breeding and crossover'])
plt.ylabel('Time per 1000 generations (μs)')
plt.xlabel('Population Size')
```
