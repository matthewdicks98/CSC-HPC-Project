```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read in the data

```python
omp_fitness = pd.read_csv("Results/omp_fitness_population.txt")
omp_fitness[' elapsed/gen_count'] = omp_fitness[' elapsed/gen_count']*1000

omp_breeding = pd.read_csv("Results/omp_breeding_population.txt")
omp_breeding[' elapsed/gen_count'] = omp_breeding[' elapsed/gen_count']*1000

omp_crossover = pd.read_csv("Results/omp_crossover_population.txt")
omp_crossover[' elapsed/gen_count'] = omp_crossover[' elapsed/gen_count']*1000
```

```python
plt.figure(figsize=(9,7))
plt.plot(omp_fitness[' pop_size'], omp_fitness[' elapsed/gen_count'],marker='o')
plt.plot(omp_breeding[' pop_size'],omp_breeding[' elapsed/gen_count'],marker='o')
plt.plot(omp_crossover[' pop_size'],omp_crossover[' elapsed/gen_count'],marker='o')
plt.legend(['fitness','fitness and breeding','fitness and breeding and crossover'])
plt.ylabel('Time per 1000 generations')
plt.xlabel('Population Size')
```
