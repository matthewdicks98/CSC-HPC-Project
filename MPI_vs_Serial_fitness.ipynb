```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mpi_particle = pd.read_csv('Results/mpi/mpi_particles.txt')
mpi_particle
```

```python
mpi_pop = pd.read_csv('Results/mpi/mpi_population.txt')
mpi_pop
```

```python
ser_particle = pd.read_csv('Results/serial/ser_particles.txt')
ser_particle
```

```python
ser_pop = pd.read_csv('Results/serial/ser_population')
ser_pop
```

## Fitness as a function of particles

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(mpi_particle[' num_particles'][:-2],mpi_particle[' final_fitness'][:-2].astype('float'),marker='o')
plt.plot(ser_particle[' num_particles'],ser_particle[' final_fitness'],marker='o')
plt.title('Fitness as a function of the number of particles')
plt.xlabel('Number of Particles')
plt.ylabel('Fitness')
plt.legend(['MPI','Serial'])
```

## Fitness as a function of population

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(mpi_pop[' pop_size'],mpi_pop[' final_fitness'].astype('float'),marker='o')
plt.plot(ser_pop[' pop_size'],ser_pop[' final_fitness'],marker='o')
plt.title('Fitness as a function of the populationsize')
plt.xlabel('Population Size')
plt.ylabel('Fitness')
plt.legend(['MPI','Serial'])
```
