```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## comapring number of particles

```python
omp_particles = pd.read_csv("Results/omp_crossover/omp_crossover_particles.txt")
omp_particles[' elapsed/gen_count'] = omp_particles[' elapsed/gen_count']*1000

mpi_particles = pd.read_csv("Results/mpi/mpi_particles.txt")
mpi_particles[' elapsed/gen_count'] = mpi_particles[' elapsed/gen_count']*1000

ser_particles = pd.read_csv("Results/serial/ser_particles.txt")
ser_particles[' elapsed/gen_count'] = ser_particles[' elapsed/gen_count']*1000


laptop_omp_particles = pd.read_csv("Results/laptop/omp_crossover/omp_crossover_particles.txt")
laptop_omp_particles[' elapsed/gen_count'] = laptop_omp_particles[' elapsed/gen_count']*1000

laptop_mpi_particles = pd.read_csv("Results/laptop/mpi/mpi_particles.txt")
laptop_mpi_particles[' elapsed/gen_count'] = laptop_mpi_particles[' elapsed/gen_count']*1000

laptop_ser_particles = pd.read_csv("Results/laptop/serial/ser_particles.txt")
laptop_ser_particles[' elapsed/gen_count'] = laptop_ser_particles[' elapsed/gen_count']*1000
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(ser_particles[' num_particles'], ser_particles[' elapsed/gen_count'],marker='D', color="teal")
plt.plot(omp_particles[' num_particles'],omp_particles[' elapsed/gen_count'],marker='o', color="violet")
plt.plot(mpi_particles[' num_particles'],mpi_particles[' elapsed/gen_count'],marker='x', color="tab:red")
plt.legend(['serial','openmp','mpi'])
plt.ylabel('Time per 1000 generations (μs)')
plt.xlabel('Number of particles')
```

```python
plt.figure(figsize=(9,7))
plt.grid(alpha=0.15)
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.25)
plt.plot(ser_particles[' num_particles'][:6], ser_particles[' elapsed/gen_count'][:6],marker='D', color="teal")
plt.plot(omp_particles[' num_particles'][:7],omp_particles[' elapsed/gen_count'][:7],marker='o', color="violet")
plt.plot(mpi_particles[' num_particles'][:7],mpi_particles[' elapsed/gen_count'][:7],marker='x', color="tab:red")

plt.plot(laptop_ser_particles[' num_particles'], laptop_ser_particles[' elapsed/gen_count'],marker='D', color="teal",dashes=[2, 2])
plt.plot(laptop_omp_particles[' num_particles'][:7],laptop_omp_particles[' elapsed/gen_count'][:7],marker='o', color="violet",dashes=[2, 2])
plt.plot(laptop_mpi_particles[' num_particles'][:7],laptop_mpi_particles[' elapsed/gen_count'][:7],marker='x', color="tab:red",dashes=[2, 2])
plt.legend(['serial (hpc)','openmp (hpc)','mpi (hpc)','serial (laptop)','openmp (laptop)','mpi (laptop)'])
plt.ylabel('Time per 1000 generations (μs)')
plt.xlabel('Number of particles')
```
